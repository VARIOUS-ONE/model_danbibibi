## 데이터 불러오기

```python
# -*- coding: cp949 -*-

naver = pd.read_csv('./data/naverReview.csv')
st_11 = pd.read_csv('./data/11stReview.csv')
```

```python
naver['star'].value_counts()
```

```python
st_11.drop(['Unnamed: 0', '상품명', '날짜', 'ID', '꾹'], axis=1, inplace=True)
st_11
```

```python
st_11['평점'].value_counts()
```

## Column rename

```python
st_11.rename(columns = {'평점':'star','리뷰':'review'},inplace=True)
st_11
```

## Concat

```python
data = pd.concat([naver, st_11], axis=0)
data
```

```python
data.iloc[199999:, :]
```

## 0과 1로 레이블 변경

```python
def changeTo01(x):
    if x<3:
        return 0
    else:
        return 1
data['star'] = data['star'].apply(changeTo01)
data
```

```python
data.to_csv('./data/review.csv')
```
